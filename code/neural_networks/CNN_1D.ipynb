```python
# Importing the necessary libraries and functions!  XGBOOST handles Naan values.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import confusion_matrix, classification_report
import os

# Function that aggregates all the csv files from both folders into two dataframes for identified and unknown
def aggregate_csvs(directory_path):
    # set up an Empty list
    dataframes = []

    # Going through directories to concatenate it all (all the CSV's)
    for subdir, _, files in os.walk(directory_path):
        for file in files:
            if file.endswith('.csv'):
                filepath = os.path.join(subdir, file)
                df = pd.read_csv(filepath)
                dataframes.append(df)

    # Combine all DataFrames into one
    aggregated_df = pd.concat(dataframes, ignore_index=True)
    return aggregated_df

identified_data = aggregate_csvs('Identified_Classifier') # target folder, identified_classifier is a split datasets
unknown_data = aggregate_csvs('Unknown_Classifier')# unknown_classifier is a split dataset



# The actual important features we want to try out.
def important_features(sigF):

    window_size_milliseconds_steering = 10000  # 10 seconds window of time
    window_size_milliseconds_distance = 1000  # 1 second window of time


    # Wheel positions features
    sigF['steering_stability'] = sigF['vtti.steering_wheel_position'].rolling(window=window_size_milliseconds_steering).std()

    # Pedal features
    sigF['brake_duration'] = sigF.groupby((sigF['vtti.pedal_brake_state'] != sigF['vtti.pedal_brake_state'].shift()).cumsum()).cumcount() * sigF['vtti.pedal_brake_state']
    sigF['brake_event'] = sigF['vtti.pedal_brake_state'].diff().apply(lambda x: 1 if x == 1 else 0) #  Brake event important!
    sigF['brake_frequency'] = sigF['brake_event'].rolling(window=window_size_milliseconds_steering).sum()


    # Lateral Position
    sigF['avg_abs_distance_from_center'] = sigF['vtti.lane_distance_off_center'].abs().rolling(window=window_size_milliseconds_distance).mean()
    sigF['lateral_velocity'] = sigF['vtti.lane_distance_off_center'].diff()
    sigF['lateral_acceleration'] = sigF['lateral_velocity'].diff()




    return sigF

identified_data_CNN = important_features(identified_data) # New dataframe with only the significant features included!
unknown_data_CNN = important_features(unknown_data)       # Same thing was the line above!



# Our significant features (the ones we defined earlier)
#significant_columns = ['steering_stability','brake_duration','brake_frequency',
                      #'avg_abs_distance_from_center','lateral_velocity','lateral_acceleration']

# New sig features

significant_columns = ['lateral_velocity','brake_frequency',
                       'avg_abs_distance_from_center','lateral_acceleration']   # WORKs

identified_data_CNN = identified_data_CNN[significant_columns] # Code that makes the dataframe shorter based on the features we want to keep
unknown_data_CNN = unknown_data_CNN[significant_columns] # Does same thing as line above
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler # Importing the necessary libraries: below is a basic version, you can make it as sophisitcated as you want
from sklearn.model_selection import train_test_split

SEQUENCE_LENGTH = 12  # This means the RNN will consider 12 rows at a time. Adjust as needed (ie: use more if you can!)

def create_sequences(data, seq_length): # Creating the actual sequences
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i + seq_length])
    return np.array(sequences)

identified_data_CNN['label'] = 1 # Setting up the labels
unknown_data_CNN['label'] = 0

# Labels and features separation (important step! )
identified_labels = identified_data_CNN.pop('label').values
unknown_labels = unknown_data_CNN.pop('label').values

# aggregating the identified and unknown data for scaling in the next line of code!
all_data_CNN = pd.concat([identified_data_CNN, unknown_data_CNN])

# Scaling the data (important step)
scaler = StandardScaler()
all_data_scaled_RNN_Bi = scaler.fit_transform(all_data_CNN)

# Separate the scaled data back into identified and unknown
identified_data_scaled_CNN = all_data_scaled_RNN_Bi[:len(identified_data_CNN)]
unknown_data_scaled_CNN = all_data_scaled_RNN_Bi[len(identified_data_CNN):]

# here, creating sequences for each class
identified_sequences_CNN = create_sequences(identified_data_scaled_CNN, SEQUENCE_LENGTH)
unknown_sequences_CNN = create_sequences(unknown_data_scaled_CNN, SEQUENCE_LENGTH)

identified_labels = identified_labels[-len(identified_sequences_CNN):]
unknown_labels = unknown_labels[-len(unknown_sequences_CNN):]

X = np.vstack((identified_sequences_CNN, unknown_sequences_CNN))
y = np.hstack((identified_labels, unknown_labels))

# Splitting (test, train etc)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python

import tensorflow as tf
from tensorflow import keras # Importing the necessary libraries (especially Keras)
from sklearn.metrics import classification_report

# Defining the 1D CNN model
model_CNN = keras.Sequential([
    keras.layers.Conv1D(32, kernel_size=3, activation='relu', input_shape=(SEQUENCE_LENGTH, len(significant_columns))),
    keras.layers.MaxPooling1D(pool_size=2),
    keras.layers.Conv1D(64, kernel_size=3, activation='relu'),
    keras.layers.MaxPooling1D(pool_size=2),
    keras.layers.Flatten(),
    keras.layers.Dense(50, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid')
])

model_CNN.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Trainining the CNN model with certain parameters, change those based on your needs and computational ability
history = model_CNN.fit(X_train, y_train, epochs=40, batch_size=50, validation_data=(X_val, y_val)) # 10 epochhs, 64 batchsize usuallyy

# Labeling the  predictions
y_pred = model_CNN.predict(X_val)
y_pred_binary = [1 if y[0] > 0.5 else 0 for y in y_pred]  # Convert probabilities to binary labels

# Classification report, Accuracy etc (just like the other models!)
report = classification_report(y_val, y_pred_binary)
print(report)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report # Hypertuning process, importing the necessary libraries!
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint # I have not ran this code yet, included it as a bonus! Might need to fix it a bit.

# Defining the CNN model (the parameters are basic, change them as needed there is lots of online documentation about this!)
model = keras.Sequential([
    Conv1D(64, kernel_size=3, padding='same', input_shape=(SEQUENCE_LENGTH, len(significant_columns))),
    LeakyReLU(),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Conv1D(128, kernel_size=3, padding='same'),
    LeakyReLU(),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Conv1D(256, kernel_size=3, padding='same'),
    LeakyReLU(),
    BatchNormalization(),
    MaxPooling1D(pool_size=2),

    Flatten(),

    Dense(512),
    LeakyReLU(),
    Dropout(0.5),

    Dense(128),
    LeakyReLU(),
    Dropout(0.5),

    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Early stopping + model checkpointing (check documentation to learn more about this)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, verbose=1, mode='min')
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True, verbose=1, mode='min')

callbacks = [early_stopping, model_checkpoint]

# Training the model (here it is trained over arbitrary epochs and other parameters). Change it as needed!
history = model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val), callbacks=callbacks)

# Loading the best model!
model = keras.models.load_model('best_model.h5')

# Predict the labels for the validation set
y_pred = model.predict(X_val)
y_pred_binary = [1 if y[0] > 0.5 else 0 for y in y_pred]

# printing the classification report etc
report = classification_report(y_val, y_pred_binary)
print(report)
```
